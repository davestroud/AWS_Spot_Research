```python
import json
import sqlalchemy
sql_url = json.load(open("sql_connection.json", "r"))['spot_db']
# Connect to the db
db = sqlalchemy.create_engine(
    sql_url,
    echo=False,
).connect()

tables = sqlalchemy.inspect(db).get_table_names()
```

```python
tables[0:5]
```

```python
import pandas as pd
import seaborn as sns
# This pulls in a table as a pd df
data = pd.read_sql_table(tables[0], con=db)

# Convert the times to real datetime objs
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.head()
```

```python
#Make them real datetimes, sort the frame
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data = data.sort_values('Timestamp', ascending=False)
data.head()
```

```python
# Get just RHEL instances
RHEL_instances = data[data['ProductDescription'] == 'Red Hat Enterprise Linux']

# Reset the index, this is stupid pandas nonsense
RHEL_instances.reset_index(drop=True, inplace=True)
RHEL_instances.head()
```

```python
RHEL_instances.shape
```

```python
# Time differences:
diffs = []
for i in range(RHEL_instances.shape[0]-1):
    diffs.append((RHEL_instances['Timestamp'][i] - RHEL_instances['Timestamp'][i+1]).total_seconds() / 60)
```

```python
sns.distplot(diffs, kde=False, rug=True)
```

```python
# Make sure you close the DB when done!
db.close()
```
